```python
df1=df.tail(80)
x=df1['Gasoline Price'].tolist()
z=df1['Diesel Price'].tolist()
y=df1.index.values.tolist()
```

```python
fig,ax=plt.subplots(1,1,figsize=[30,10])

ax.set_title('British Petroleum (Poland)')
ax.set_xlabel('Year')
ax.set_ylabel('Gasoline Price')

ax.plot(y,x,color='green',label="Gasoline Price",linewidth=2.0,marker='+',markersize=8)
ax.plot(y,z,color="#4c4959",label="Diesel Price",linewidth=2.5,marker='s',markersize=8)

plt.grid(True)
plt.xticks(rotation=50)
fig.autofmt_xdate()
plt.tight_layout()

ax.legend(loc=4)
ax.legend()

fig.savefig("filename.png", dpi=200)
fig.savefig("filename.pdf", dpi=200)
```
